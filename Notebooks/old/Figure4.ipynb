```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_50_100=pd.read_csv('../analysis_files/df_length_50_100.csv')
df_50_100['len'] = df_50_100['len'].str.replace('_', '-')
df_50_100.head(10)
```

```python
df_100_500=pd.read_csv('../analysis_files/df_length_100_500.csv')
df_100_500['len'] = df_100_500['len'].str.replace('_','-')
df_100_500.head()
```

```python
df_500_1000=pd.read_csv('../analysis_files/df_length_500_1000.csv')
df_500_1000['len'] = df_500_1000['len'].str.replace('_','-')
df_500_1000.head()
```

```python
df_1000_=pd.read_csv('../analysis_files/df_length_1000_.csv')
df_1000_['len'] = df_1000_['len'].str.replace('_','+')
df_1000_.head()
```

```python
all = pd.concat([df_50_100, df_100_500,df_500_1000,df_1000_], axis=0)
all.tail(77)
```
